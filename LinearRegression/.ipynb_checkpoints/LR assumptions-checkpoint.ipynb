**The Assumptions**
1. Linear relationship between input and output
2. No Multicolinearity
3. Normality of residuals
4. Homoscedasicity
5. No auto co relation of errors

```python
# lets work with a dataset with building the linear regression model then we will see how the 
# assumptions are taking place
# let's import all necessary libraries
import pandas as pd 
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_diabetes
```

```python
diabetes = load_diabetes()
```

```python
diabetes.keys()
```

```python
x = pd.DataFrame(data = diabetes.data, columns = diabetes.feature_names)
```

```python
x.head()
```

```python
y = pd.DataFrame(data = diabetes.target, columns = ['target'])
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.3,random_state=1)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model.fit(xtrain,ytrain)
```

```python
ypred = model.predict(xtest)
residual = ytest - ypred
```

**Now lets find out the truth behind all the assumptions**

**Linear relationship between input and output**

If we plot this data into a plot then we will be able to see that the relationship between 
the features are linear in nature . We will compare every independent variable with the target variable to understand if the relationship is linear or not. We will use scatterplot to identify the same.


```python
plt.scatter(x['age'],y['target'])
```

```python
plt.scatter(x['sex'],y['target'])
```

```python
plt.scatter(x['bmi'],y['target'])
```

```python
plt.scatter(x['bp'],y['target'])
```

**Multi Co-Linearity**

This simply states that the independent variables present in our data are not co related. That means for example we have a diabetes dataset with us. In this set age , sex bmi etc are independent variables , and this assumption says that these variables should not be highly co related , that means if age changes that doesn't mean the value places in sex column will change based on that.

Let's find the same with the help of variance inflation factor.

```python
'''Importing the vif module from statsmodel'''
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
'''If the vif for all independent variable comes near 1 then there is less or no multi co linearity , but
if it's 5 and above then there is high chances of multi co linearity'''
vif = []
for i in range(xtrain.shape[1]):
    vif.append(variance_inflation_factor(xtrain,i))
pd.DataFrame({'vif':vif},index=x.columns[0:10]).T
```

**Normal Distribution of Residuals**

Residuals are nothing but the errors , which means the difference of actual and predicted value. And this assumption states that this residuals will be normally distributed , mean will be closed to 0 and few data points will be spread far.

```python
''' Let's plot the residuals into a kde plot from seaborn which will help us visualize the distrinution '''
sns.displot(residual,kind ='kde')
```

**Homoscedasticity**

This assumption states that the spread of residualsare uniform in nature and with the help of scatterplot we should be able to identify the same. If in any case the spread is not uniform , that's called as a case of heteroscedasticity.

```python
plt.scatter(ypred,residual)
plt.show()
```

**No Auto Co Relation Between Residuals**

This assumption simply states that among the residuals , no auto co relation can be present , which means , if we plot the residuals there won't be any pattern visible.

```python
plt.plot(residual)
```
